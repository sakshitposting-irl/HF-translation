__This notebook will serve as our base to learn language translation using deep learning.__

_This notebook is an introduction to huggingface transformers library which is a giant open source community dedicated to NLP research._

We begin by importing the necessary libraries.

```python
import gradio as gr 
from transformers import pipeline
```

```python
translator = pipeline("translation", model="vsrinivas/marian-finetuned-kde4-en-to-hi")
result = translator('She say do you love me, I tell her only partly')
print(result[0]['translation_text'])
```

```python
print(result[0]['translation_text'])
```

```python
def translation_function(text):
    result = translator(text)
    return result[0]['translation_text']
```

```python
gr.Interface??
```

```python
interface = gr.Interface(fn = translation_function,
                         inputs = gr.components.Textbox(lines=2, placeholder='Text Translator'),
                         outputs='text')
```

```python
interface.launch(share=True)
```
